```python
import re
import numpy as np
import itertools
from sklearn.feature_extraction.text import CountVectorizer
data=[]
with open('/home/justin/Desktop/Jupyter/Test Data/Text/Washington.txt', 'r') as file:
    data =file.read()
#lists
W_lines_data=re.findall(r"(?<=(\d\s))(.+)",data)
W_words_per_line_data = [line[1].rstrip().split(" ") for line in W_lines_data]
W_total_words_data = list(itertools.chain(*W_words_per_line_data))
W_features_data = CountVectorizer().fit(raw_documents = W_total_words_data).get_feature_names()

#connecting words that are not connected
W_total_full_words_data = W_total_words_data.copy()
a =0
for word in W_total_full_words_data:
    if word.endswith("-"):
        W_total_full_words_data[a]=W_total_full_words_data[a][:-1] + W_total_full_words_data[a+1]
        W_total_full_words_data.pop(a+1)
        
    if word.startswith("-"):
        W_total_full_words_data[a-1]=W_total_full_words_data[a-1] + W_total_full_words_data[a][1:]
        W_total_full_words_data.pop(a)
    a +=1
```

```python
print("Total number of raw words:                               ",len(W_total_words_data))
print("Total number of complete words                           ",len(W_total_full_words_data))
print("Total number of features:                                ",len(W_features_data))
```

## Prototype

```python
G_total_words_data #this data still has empty block characters
print(G_total_words_data)
W_total_full_words_data
n =10

#split is number of words per chunk
def splitter(raw_words, split):
    grouped_words=[]
    for x in raw_words:
        if(len(raw_words)>=10):
            grouped_words.append(" ".join(raw_words[:split]))
            raw_words= raw_words[split:]
            
    grouped_words.append(" ".join(raw_words))
    return grouped_words

G_X= splitter(G_total_words_data.copy(),10)
G_y= [0]*len(G_X)
W_X= splitter(W_total_full_words_data.copy(),10)
W_y= [1]*len(W_X)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
y_train=G_y + W_y
vect = CountVectorizer(min_df=5,stop_words="english").fit(G_X+W_X)
X_train = vect.transform(G_X+W_X)

```

### Multinomial Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score 
scores = cross_val_score(MultinomialNB(),X_train,y_train, cv=10)
print("Mean cross-validation accuracy: {:.2f}".format(np.mean(scores)))
```

### Bernoulli  Naive Bayes

```python
from sklearn.naive_bayes import BernoulliNB
scores = cross_val_score(BernoulliNB(),X_train,y_train,cv=10)
print("Mean cross-validation accuracy: {:.2f}".format(np.mean(scores)))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
scores = cross_val_score(LogisticRegression(),X_train,y_train,cv=10)
print("Mean cross-validation accuracy: {:.2f}".format(np.mean(scores)))
```
